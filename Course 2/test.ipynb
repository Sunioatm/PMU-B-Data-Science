```python
import matplotlib.pyplot as plt
import matplotlib.figure as figure
```

```python
dataX = ["1 Oct 66", "3 Oct 66", "4 Oct 66", "6 Oct 66", "7 Oct 66", "11 Oct 66", "17 Oct 66", "21 Oct 66", "31 Oct 66"]
dataE20 = [37.14, 36.84, 36.44, 35.94, 35.44, 35.84, 35.54, 35.94, 36.24]
dataE85 = [36.79,36.49,36.09,35.59,35.59,35.99,35.69,36.09,36.39]
```

```python
# plot continuous line
fig = plt.figure(figsize=(10, 5), facecolor="lightblue", edgecolor="black", linewidth=4)
ax = fig.add_axes([0, 0, 0.8, 0.8]) # main axes
ax1 = ax.plot(dataX, dataE20, 'r', label='E20')
ax2 = ax.plot(dataX, dataE85, 'b', label='E85')

ax.set_title("E20 vs E85", fontsize=20)
ax.set_xlabel("Date", fontsize=16)
ax.set_ylabel("Price", fontsize=16)

plt.show()
```
